 ## Monty Hall Problem - Gambling Game
    
    Suppose you are on a game show, and you have the choice of picking one of three doors: Behind one door is a car; behind the other doors, goats. You pick a door, let’s say door 1, and the host, who knows what’s behind the doors, opens another door, say door 3, which has a goat. The host then asks you: do you want to stick with your choice or choose another door? [1]

    Is it to your advantage to switch your choice of door?

Based on probability, it turns out it is to our advantage to switch the doors    

```python
import random
import matplotlib.pyplot as plt
```

```python
def mc_simulation(inital_state,num_simulation):
    switch , no_switch = 0,0 # switch win , no_switch win
    switch_probability = []
    no_switch_probability = []
    for i in range(num_simulation):
        random.shuffle(inital_state)
        c = random.randrange(2)
        if inital_state[c]!='car':
            switch+=1
        else:
            no_switch+=1
        
        switch_probability.append(switch/(i+1))
        no_switch_probability.append(no_switch/(i+1))
        
        if i==num_simulation-1: #add label just for the last plot
            plt.plot(switch_probability,label='switch')
            plt.plot(no_switch_probability,label='no switch')
            
        else:
            plt.plot(switch_probability)
            plt.plot(no_switch_probability)
        
    plt.axhline()
    plt.xlabel('num_simulation')
    plt.ylabel('probability of win')
    plt.title('Monte Carlo switch vs no-switch win')
    plt.legend()
    print('Probability of win if switched : ',switch_probability[-1])
    print('Probability of win if not switched : ',no_switch_probability[-1])
```

```python
inital_state_1 = ['car','goat','goat']
mc_simulation(inital_state_1,100)
```

```python
inital_state_2 = ['goat','car','goat']
mc_simulation(inital_state_2,1000)
```

```python
inital_state_3 = ['goat']*3+['car']*2
mc_simulation(inital_state_3,1000)
```
